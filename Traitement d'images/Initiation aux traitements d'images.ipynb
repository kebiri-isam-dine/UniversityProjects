# Initiation aux traitements d'images 

```python
from PIL import Image
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
imgpil = Image.open("C:\\Users\\moham\\OneDrive\\Bureau\\Lenna.png") #ouvrir et conserver l'image dans imgpil
plt.imshow(imgpil) #afficher l'image
```

```python
img = np.array(imgpil) # Transformation de l'image en tableau numpy (matrice)
```

```python
img.shape #on utilise image.shape pour obtenir la taille le long de chaque dimension
```

```python
img[1,1] #pour avoir la couleur du pixel
```

```python
print(img[::3]) #affichage de la matrice de l'image
```

```python
#extraction des matrices red green blue
red=img[:,:,0]
print(red)
green=img[:,:,1]
print(green)
blue=img[:,:,2]
print(blue)
```

```python
img = imgpil.convert('LA') #fonction predifini pour convertir image en niveau de gris
img.save('C:\\Users\\moham\\OneDrive\\Bureau\\Lennagrey.png') # save image en niveau de gris
#affichage image en niveau de gris
plt.imshow(img)
```

```python
# passer du domaine spatial au domaine frequentiel en appliquant la fft  2 dimensions
imgf=np.fft.fft2(img) #calcul de la fft
imgabs=np.abs(imgf) #calcul de l'amplitude (partie relle)
```
